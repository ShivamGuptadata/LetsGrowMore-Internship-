# Introduction
Read the image in RBG format and then convert it to a grayscale image. This will turn an image into a classic black and white photo. Then the next thing to do is invert the grayscale image, this will be our inverted grayscale image. Inversion can be used to enhance details. Then we can finally create the pencil sketch by mixing the grayscale image with the inverted blurry image. This can be done by dividing the grayscale image by the inverted blurry image. 

##Import Required Libraries

```python
import cv2
import pandas as pd
import matplotlib.pyplot as plt
from skimage import io
from google.colab.patches import cv2_imshow
%matplotlib inline
```

```python
#Read the Image
img=cv2.imread('/labradoe-outside.jpg')
```

```python
# Display the Image
cv2_imshow(img)
```

```python
#Resize the image
(height,width)=img.shape[:2]
new_image=cv2.resize(img,(int(width/2),int(height/2)), interpolation=cv2.INTER_AREA)
cv2_imshow(new_image)
```

#Make function to convert real image into pencil sketch image

```python
def convert_into_pencil(image):
  rgb=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)              # Change into RGB format
  final_img=cv2.hconcat((image,rgb))
  show1=cv2_imshow(final_img)
  print("RGB format:",show1)
  gray_img=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)        # Change into classic black & white photo
  print("Gray colour format:",cv2_imshow(gray_img))
  invert_gray=cv2.bitwise_not(gray_img)                  # Invert the grayscale image
  print("Inverted Gray colour format:",cv2_imshow(invert_gray))
  blur_img=cv2.GaussianBlur(invert_gray,(21,21),0)       # Blurry image
  print("Blurry Image:",cv2_imshow(blur_img))
  invert_blur=cv2.bitwise_not(blur_img)                  # Inverted Blurry image
  print("Inverted Blurry Image:",cv2_imshow(invert_blur))
  pencil_img=cv2.divide(gray_img,invert_blur,scale=256.0) # Pencil-Sketch image
  
  return print("Pencil sketch Image:",cv2_imshow(pencil_img)) 
```

```python
#Display the final image
convert_into_pencil(new_image)
```
